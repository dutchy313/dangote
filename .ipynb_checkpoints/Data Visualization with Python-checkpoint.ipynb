```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim
```

```python
dangote = pd.read_csv('petroleum.csv')
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
def get_lat_lon(country):
    try:
        location = geolocator.geocode(country)
        return location.latitude, location.longitude
    except:
        return None, None
```

```python
dangote['Latitude'], dangote['Longitude'] = zip(*dangote['Country'].apply(get_lat_lon))
```

```python
dangote_clean = dangote.dropna(subset=['Latitude', 'Longitude'])
```

```python
print (dangote)
```

```python
political_countries_url  = 'http://geojson.xyz/naturalearth-3.3.0/ne_50m_admin_0_countries.geojson'
```

```python
m = folium.Map(location=[30, 10], zoom_start=3,tiles="cartodb positron")
```

```python
folium.Choropleth(
    geo_data=political_countries_url,
    data=dangote,
    columns=['Country', 'Amount'],
    key_on='feature.properties.name',
    fill_color='RdYlGn_r',
    fill_opacity=0.7,
    line_opacity=0.3,
    nan_fill_color="white",
    nan_fill_opacity=0.2,
    legend_name='Amount (in millions USD) of Refined Petroleum Imported to Nigeria in 2022',
).add_to(m)
```

```python
for _, row in dangote.iterrows():
     folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        tooltip=f"{row['Country']}: {row['Amount']}",
        icon=folium.Icon(color='blue', icon='star')
    ).add_to(m)
```

```python
m.save('petroleum.html')
```
